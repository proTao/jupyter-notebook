```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
dir="/home/protao/tao/output"


def walkAllFile(dir, suffix, func):
    "对dir文件夹下，满足suffix后缀的文件，执行func函数"
    for filename in os.listdir(dir):
        filepath = os.path.join(dir, filename)
        if os.path.isdir(filepath):
            walkAllFile(filepath, suffix, func)
        else:
            if filename.endswith(suffix):
                func(filepath)

def transTime(filepath):
    path="/home/protao/tao/output/domain/"
    filename = filepath.split("/")[-2]
    print(filename)
    x=[]
    y=[]
    z=[]
    with open(filepath) as f:
        lines = f.readlines()
    for line in lines:
        splited = line.split("\t")
#         print(splited)
        x.append(splited[0])
        y.append(splited[1].strip())
        z.append(splited[2].strip())
    
#     print(x)
    x = list(map(lambda x: int(x), x))
    y = list(map(lambda x: int(x), y))
    z = list(map(lambda x: int(x), z))
    
#     print(x)
    x = list(map(lambda x: x-822511, x))
#     print(x)
    continuous_x=[]
    continuous_y=[]
    continuous_z=[]
    for i in range(len(x)-1):
#         print(i,y[i])
        index = int(x[i])
        next_index = int(x[i+1])
        delta = next_index - index
        continuous_x.append(int(index))
        continuous_y.append(int(y[i]))
        continuous_z.append(int(z[i]))
        while delta>1:
            index += 1
            continuous_x.append(index)
            continuous_y.append(0)
            continuous_z.append(0)
            delta -= 1
    continuous_x.append(int(x[-1]))
    continuous_y.append(int(y[-1]))
    continuous_z.append(int(z[-1]))
#     print(continuous_y)
    start = continuous_x[0]
    
    result_x = []
    result_y = []
    result_z = []
    for i in range(start):
        result_x.append(i)
        result_y.append(0)
        result_z.append(0)
    result_x.extend(continuous_x)
    result_y.extend(continuous_y)
    result_z.extend(continuous_z)
    for i in range(len(result_x), 1488):
        result_x.append(i)
        result_y.append(0)
        result_z.append(0)
#     print(result_y)
    with open(path+filename+"/data", 'w') as fo:
        for i in range(len(result_x)):
            fo.write(str(result_x[i]))
            fo.write("\t")
            fo.write(str(result_y[i]))
            fo.write("\t")
            fo.write(str(result_z[i]))
            fo.write("\n")
        
    
def secondToDate(second):
    time = second // 2
    hour = time % 24
    time = time // 24
    date = time % 30
    time = time // 30
    month = time % 12
    return str(month)+"月"+str(date)+"日"+str(hour)+"时"
                
def drawFlow(file_path):
    with open(file_path) as f:
        lines = f.readlines()

    x=[]
    y=[]
    z=[]
    for line in lines:
        splited = line.split("\t")
        x.append(splited[0])
        y.append(splited[1].strip())
        z.append(splited[2].strip())
    
#     date = list(map(secondToDate, continuous_x))

    fig = plt.figure(file_path, figsize=(15,5), dpi=80)
    ax = fig.add_subplot(1,1,1)
    plt.plot(x, y, 'b-', lw=1)
    plt.plot(x, z, 'r-', lw=1)
    plt.show()
    filename = file_path +".pdf"
    fig.savefig(filename, bbox_inches='tight')
    
# transTime("/home/protao/tao/output/domain/cloudfront.net/part-r-00000")
# drawFlow("/home/protao/tao/output/domain/cloudfront.net/data")
walkAllFile("/home/protao/tao/output/domain", "00000", transTime)
walkAllFile("/home/protao/tao/output/domain", "data", drawFlow)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

length = 400
convergence = 60
convergence_line=[60 for i in range(length)]
x=[i for i in range(length)]

y1=[10]
alpha = 0.03
for i in range(1, length):
    new_y = y1[i-1] + (convergence - y1[i-1]) * alpha
    random = (np.random.random()-0.5)*0.3
    y1.append(new_y + random)

y2=[30]
alpha=0.015
for i in range(1, length):
    new_y = y2[i-1] + (convergence - y2[i-1]) * alpha
    random = (np.random.random()-0.5)*0.3
    y2.append(new_y + random)

y3=[95]
alpha=0.1
for i in range(1, length):
    new_y = y3[i-1] + (convergence - y3[i-1]) * alpha
    random = (np.random.random()-0.5)*0.3
    y3.append(new_y + random)    

fig = plt.figure("thread", figsize=(15,7), dpi=80)
ax = fig.add_subplot(1,1,1)
ax.plot(x,y1,"r-",lw=2,label="threat value")
ax.plot(x,y2,"g-",lw=2,label="threat value")
ax.plot(x,y3,"b-",lw=2,label="threat value")
ax.plot(x,convergence_line, "r--",lw=1)
plt.show()
```
