# ndarray对象的内部机理

## 存储方式

- 所有的的数组对象都是数据块的一个跨度视图（strided view），跨度是切片操作不执行复制只返回视图的原因所在
- 对象包括：
    - 一个指向数组（系统内存块）的指针
    - 数据类型或者dtype
    - shape
    - stride，表示前进到当前维度下一个元素所需要跨过的字节数，例如一个shape是(3,4,5)的C顺序的float64数组，stride为(160,40,8)，而且跨度可以使负数用来表示反序对象。
    
## 数据结构体系
- 主要介绍两个方法：issubdtype和mro，前者用于判断对象数据类型是否属于某个超类，后者返回某种数据类型所有的父类。
- numpy中的数据类型转换，不能直接改原数据的dtype！只能用函数astype()。

```python
import numpy as np
```

```python
a = np.arange(3*4*5).reshape(3,4,5)
a.strides
```

```python
print(a.dtype)
print(np.issubdtype(a.dtype, np.integer))
```

```python
np.float.mro()
```

# 高级数组操作

## 数据重塑
- reshape
- 将多维数组转化为一维数组：flattening（扁平化），reval（散开）
    - flatern函数返回源数据的副本，ravel返回源数据的视图
- 关于C和Fortran顺序
    - 由于某些历史原因，行和列优先顺序又分别被称为C和Fortran顺序
    - 行优先顺序是指每行中的数据项被存放到相邻内存位置，列优先顺序类似
    - reshape或者ravel函数接受order参数来表示数组数据优先存放位置，可以是‘C’或者‘F’，还有其他可选参数，详见文档
    - 原则上看C优先顺序先处理更“里面”的维度（axis号更大），Fortran相反

```python
a = np.arange(24).reshape(4,6)
a
```

```python
a.ravel('F')
```

## 数组的合并和拆分
- np.concatenate，输入需要连接的列表，可以指定轴
- 对于常见的连接操作，np提供便捷方法vstack和hstack
- np.split执行连接的相反操作，传入待切分的数组和切分点列表和指定轴
- 堆叠辅助类：r_和c_（详见手册吧。。。）


|函数|说明|
|:----|:---|
|concatence|常规连接|
|vstack, row_stack|在0轴上|
|hstack|在1轴上|
|column_stack|与hstack的区别是如果输入的是一维数组，会先将其转化为二维列向量|
|dstack|在轴2上|
|split|常规拆分|
|hsplit, vsplit, dsplit|沿0,1,2轴的split快捷方式|

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

print("vstack / row_stack")
print(np.vstack((a,b)))

print("hstack")
print(np.hstack((a,b)))

print("column_stack")
print(np.column_stack((a,b)))
```

```python
a = np.arange(20).reshape(10,2)
a1, a2, a3 = np.split(a, [5,8], axis=0)
print(a1)
print(a2)
print(a3)
```

## 重复操作
- 其实由于numpy的广播机制，需要手动重复数组的操作比较少
- repeat是元素级别的重复
- tile是对象整体的重复

```python
a = np.arange(6).reshape(2,3)

# 沿着0轴，第一个元素重复3次，第二个元素重复2次
a.repeat([3,2], axis=0)
```

```python
# 把a看成一个单元，重复成一个3*2的矩阵
np.tile(a, [3,2])
```

## 花式索引的等价函数：take和put
- ndarray有两个方法专门用于获取和设置单个轴向上的选区
- put不接受axis函数，只在数组的扁平化一维版本上进行索引，take在不设置axis时也是采用扁平化索引
- take/put貌似比花式索引的效率高不少

```python
a = np.arange(20).reshape(4,5)

# 默认情况下使用扁平化索引
print(a.take([1,2,3]))

# put只能使用扁平索引,可以使用一个大小相同的数组一一赋值或者传递一个标量利用广播机制整体赋值
a.put([1,2,3], [-1, -2, -3])

# 指定轴的时候类似于花式索引
print(a.take([1,2,3], axis=1))
```

# 广播
- **原则：如果两个数组的后缘维度（trailing dimension，即从末尾开始算起的维度）的轴长一致或者其中一方是1，则认为他们是广播兼容的。广播会在缺失与轴长为1的维度上进行。**
- 通过np.newaxis新增维度



```python
# 这个简单的例子可以说明使用广播时需要注意的地方

a = np.arange(12).reshape(3,4)
print(a)
print(a.shape)
print()

print(a.mean(0))
print(a.mean(0).shape)
```

```python
a-a.mean(0)
```

```python
print(a.mean(1))
print(a.mean(1).shape)

# 如果a-a.mean(1)就会报错，因为不符合前面的原则，后缘维度不一致
```

```python
# 为了解决后缘维度不一致的问题，(3,4)形状的数组不能广播(3,)的数组，但是可以广播(3,1)形状的数组
# 通过np.newaxis新增维度

m = a.mean(1)[:,np.newaxis]
print(m)
print(m.shape)

a-m
```

# ufunc高级应用
- numpy的每一个二元ufunc都有一些用于执行特定矢量化操作的特殊方法

|方法|说明|
|:----|:----|
|reduce(x)|通过连续执行原始运算的方式对值进行求和|
|accumulate(x)|累计运算，保留每次reduce结果|
|reduceat(x,bins|分组后进行reduce|
|outer(x,y)|x和y中每个元素都与另外的对象中的每个元素参与一次运算|

```python
a = np.array([1,2,3,4,5])
np.add.reduce(a)
```

```python
np.add.accumulate(a)
```

```python
# 类似与split，输入拆分点
# 拆分点列表中第一个值之前的数据不考虑，比如这里输入的是[1,3],那么计算的分组就是[1,3)和[3,end)
np.add.reduceat(a,[1,3])
```

```python
np.add.outer(a, np.array([1,2,3]))
```

## 自定义ufunc
- np.frompyfunc接受一个python函数以及两个分别表示输入输出参数数量的整数，用frompyfunc创建的函数总是返回python object数组
- np.vectorize类似于frompyfunc，但是在类型推断上更智能一些
- 有预计算每个元素是都是要执行一次python函数调用，因此比numpy自带的基于c的ufunc更慢

```python
def add_element(x,y):
    return x + y

add_them = np.frompyfunc(add_element, 2, 1)
add_them([1,2,3,4,5], [5,4,3,2,1])
```

```python
add_them = np.vectorize(add_element, otypes=[np.int16])
add_them([1,2,3,4,5], [5,4,3,2,1])
```

# 结构化和记录式数组
- 默认的ndarray是一种同质化容器，也就是说，在他表示的内存块中个元素占用的字节数相同，结构化数组是一种特殊的ndarray，类似于c语言中的struct
- 通过元组方式设置名称与数据类型，通过列表将其组合与嵌套
- 通过设置的name可以访问struct中所有元素的指定字段，以列表形式返回数据视图

```python
dtype = [('x', np.float64), ('y', np.int32)]

struct = np.array([(1.1, 1), (2.2, 2)], dtype=dtype)

print(struct[0])

print(struct['x'])
```

```python
# 设置大小, 如下面的方式设置，x就是长度为3的数组
dtype = [('x', np.float64, 3), ('y', np.int32)]
a = np.zeros(4, dtype=dtype)
print(a)
```

```python
# 可以进行嵌套
dtype = [('x', [('x1', 'f8'), ('x2', 'f4')]), ('y', np.int32)]
a = np.array([((1, 2), 5),((3, 4), 6)], dtype=dtype)

a['x']
```

# 间接排序
- argsort返回数组排序后个元素对应的索引
    -可以接受三种排序算法：quicksort/ mergesort/ heapsort
- lexsort和argsort差不多它可以一次性对多组键进行间接排序，传入二维列表，每一维都是一组键，注意后面的排序键优先级更高
- searchsorted执行二分搜素进行元素插入

```python
# example 1
a = np.array([5,0,1,3,2])
indexer = a.argsort()
a[indexer]
```

```python
# example 2 : 根据某一行排序二维数组
a = np.random.randn(3,5)
a[0] = [5,0,1,3,2]
print(a)

print("sort by line 1")
a[:, a[0].argsort()]
```

```python
# example 3: 通过mergesort进行稳定排序
value = np.array(['3:1','2:1','2:2','3:2','2:3','1:1','3:3','1:2'])
key = np.array([3,2,2,3,2,1,3,1])

print("堆排序，不稳定")
print(value[key.argsort(kind="heapsort")])

print("合并排序，稳定")
print(value[key.argsort(kind="mergesort")])
```

```python
# example 4：通过多组键进行排序
key1 = np.array([1,2,3,1,2])
key2 = np.array([1,1,1,2,2])

indexer = np.lexsort((key2, key1))
for i in zip(key1[indexer], key2[indexer]):
    print(i)
```

# 高级数组输入输出
- 内存映像文件
    - 内存映像文件是一种将磁盘上的非常大的二进制数据文件当做内存中的数组进行处理的方式
    - np的memmap对象， 允许将大文件分成小段进行读写，而不是一次性读入内存，memmap拥有和普通数组一样的方法
    - 将数据赋值给视图时，数据会先被缓存在内存中，利用flush可以将其写入磁盘
    - 可以将内存映像直接看作是存放在磁盘上的ndarray

# 连续内存
- ndarray是否存储于连续的内存块对于性能的影响很大，默认情况下numpy数组以C型连续的方式存储。
- 使用flags属性查看这些信息


```python
a= np.ones((10,10), order='C')
a.flags
```

```python
a.flags.contiguous
```

```python
# 构造视图可能会破坏这种切连续性
print(a[:5].flags)
print()
b = a[:,:5]
print(b.flags)
```

```python
# 通过copy还原连续性
c = b.copy("C")
print(c.flags)
```
